```python
# Melakukan import data dari file excel (.xlsx) dan melihat sekilas isi data
import pandas as pd 
db = pd.read_excel('C:/Users/AsusUser1827/Documents/ESA TAURAN/DATASET/FIFA21 - Insight Pemain Sesuai Kriteria/Career Mode player datasets - FIFA 15-21.xlsx')
db.head()
```

```python
#Mencari kiper dengan tinggi badan minimal 180 dan berkebangsaan Germany dengan umur tidak lebih dari 25 tahun sebagai kiper cadangan dan ranking liganya maks 3 

kiper_muda = db[(db['height_cm'] >180) & (db['nationality'] == 'Germany') & (db['age'] <= 25) & (db['player_positions'] == 'GK') & (db['league_rank'] <= 3) & (db['goalkeeping_reflexes']) & (db['goalkeeping_kicking'] >70) & (db['club_name'] !='Manchester City')]
kiper_muda.sort_values(['age'], ascending = True)  
kiper_muda[['long_name']]

# Dari proses filter, maka kita menemukan pemain sesuai dengan kriteria yang diperlukan oleh Manchester United dengan umur 
```

```python
#Mencari Bek dengan kualitas kontrol bola yang baik, mau berapapun umurnya. 

bek_kontrol = db[(db['player_positions'].str.contains('CB')) & (db['skill_ball_control'] >80) & (db['club_name'] !='Manchester City')]
bek_kontrol[['long_name', 'skill_ball_control', 'release_clause_eur']]
bek_kontrol[['long_name']]
```

```python
#Mencari Gelandang Bertahan dengan visi bermain, teknik intercept, dan skill long-passing yang baik

gelandang_bervisi = db[(db['player_positions'].str.contains('CDM')) & (db['mentality_vision'] >80) & (db['mentality_interceptions'] >80) & (db['skill_long_passing'] >80) & (db['club_name'] !='Manchester City')]
gelandang_bervisi[['long_name']]
```

```python
# Mencari striker muda dengan kriteria umur dibawah 21 tahun, memiliki umpan / crossing yang baik, skill dribbling yang baik, dan stamina yang sangat baik 

striker_muda = db[(db['player_positions'].str.contains('RW', 'LW')) & (db['age'] <=21) & (db['attacking_crossing'] >75) & (db['skill_dribbling'] >75) & (db['power_stamina'] >75)]
striker_muda
```
